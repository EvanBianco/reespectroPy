```python
import glob
import numpy as np
import pandas as pd
from docx.api import Document
from spec_io import get_OF7923_meta, get_OF7923_data
```

```python
%load_ext autoreload
%autoreload 2
```

```python
min_dir = '../input_data/of_7923/Minerals/'
min_folders = glob.glob(min_dir + '*')
min_folders = [folder.split('/')[-1] for folder in min_folders]
spectra_files = glob.glob(min_dir + '**/*.txt', recursive=True)
spectra_meta_files = glob.glob(min_dir + '**/*.docx', recursive=True)
```

```python
data_and_meta_files = []
for d, m in zip(spectra_files, spectra_meta_files):
    data_and_meta_files.append([d, m])
```

```python
# build_data_frame
data = []
count = 0
for datafile, metafile in data_and_meta_files:
    meta = get_OF7923_meta(metafile)
    try:
        header, wvls, refl = get_OF7923_data(datafile)
        data.append(list(meta.values()) + list(refl))
    except:
        pass
    count += 1
       
# build the column names from the last file.
df_columns = list(meta.keys()) + list(wvls.astype(int))

data = np.array(data)
data_nums = data[:,10:].astype(float)
df = pd.DataFrame(data=data, columns=df_columns)
df.shape
```

```python
# A look at the meta data fields
df.columns.values[:15]
```

```python
label_dict = {}
with open('../input_data/mineral_to_label_dict.csv') as f:
    for line in f:
        try:
            k, v = line.strip().split('\t')
            label_dict[k] = v
        except:
            k, v = line.strip().split(';')
            label_dict[k.strip()] = v.strip()
```

```python
# Do we have and REE bearing minerals? Should be no.
set(df['Mineral_Name'].map(label_dict))
```
