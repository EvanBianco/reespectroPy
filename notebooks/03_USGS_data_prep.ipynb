```python
data_dir = '../input_data/USGS_Speclib_07a/USGS Speclib07 REE Phase'
spectra_files = glob.glob(data_dir + '/*.txt', recursive=True)
```

```python
omit_minerals = ['Forsterite']

data = []
count = 0
for i, datafile in enumerate(spectra_files):
    folder_id, mat_name, sample_id, instr_info, _ = spec_io.get_USGS_data_header(datafile)
    meta = {'Mineral_Name': mat_name, 'ID': sample_id, 'instr_info': instr_info}
    header, wvls, ref = spec_io.get_USGS_data(datafile, encodings='us-ascii')
    if mat_name in omit_minerals:
        print('skipping:', sample_id, mat_name)
    else:
        data.append(list(meta.values()) + list(ref))

df_columns = list(list(meta.keys()) + list(wvls.astype(int)))
data = np.array(data)
df = pd.DataFrame(data=data, columns=df_columns)
```

```python
# Get mineral labels, and mineral labels as lists
with open('../input_data/mineral_label_encoding.csv') as f:
    lines = f.readlines()
min_names, label_idx = [], []

for line in lines[1:]:
    if line.count(',') >= 2:
        # there will be some loss in the middle
        a = line.strip().split(',')[:-1]
        b = line.split(',')[-1]
        if len(a) == 2:
            a = a[0] + ',' + a[1]
        if len(a) == 3:
            a = a[0] + ',' + a[1] + ',' + a[2]
        
        print(f'watch out for \n  {a}')
    else:
        a, b = line.strip().split(',')
    min_names.append(a)
    label_idx.append(int(b.strip()))
    
len(min_names), len(label_idx)
```

```python
# see if we are missing any minerals
for i, name in enumerate(df['Mineral_Name']):
        if not name in min_names:
            print(name)
```

```python
# Assign labels
label_dict = {}
with open('../input_data/mineral_to_label_dict.csv') as f:
    for line in f:
        try:
            k, v = line.strip().split('\t')
            label_dict[k] = v
        except:
            print(line)
            k, v = line.strip().split(';')
            print(k, v)
            label_dict[k.strip()] = v.strip()
```

```python
df['cm_ree_labels'] = df['Mineral_Name'].map(label_dict)

# Move the last column to the first position
cols = df.columns
cols = list(cols[-1:]) + list(cols[:-1])
df = df[cols]
```
